```python
import numpy as np
import pandas as pd

df = pd.read_csv('order_brush_order.csv')
df
```

```python
df = df[df[['orderid','shopid','userid','event_time']].notnull()] #delete the null
df
```

```python
df['year'] = pd.DatetimeIndex(df.event_time).year
df['month'] = pd.DatetimeIndex(df.event_time).month
df['day'] = pd.DatetimeIndex(df.event_time).day
df['hour'] = pd.DatetimeIndex(df.event_time).hour
df['mins'] = pd.DatetimeIndex(df.event_time).minute
print(df)
```

```python
shopcount = df.groupby(['shopid'],as_index=False)['shopid'].agg({'cnt':'count'}) #as_index:let shopid not to be index
shopcount
```

```python
df1 = shopcount[shopcount['cnt'] > 2].reset_index(drop=True)
df1
```

```python
del df1 ['cnt']
```

```python
df2 = pd.merge(df,df1, left_on='shopid', right_on='shopid', how='inner') 
df2
```

```python
df2.sort_values(by=['shopid','event_time']).reset_index(drop=True)
```

```python
usercount = df.groupby(['userid'],as_index=False)['userid'].agg({'cnt':'count'}) #as_index:讓shopid不要變成index
usercount
```

```python
df_user = usercount[usercount['cnt'] > 2].reset_index(drop=True)
df_user
```

```python
del df_user ['cnt']
```

```python
df3 = pd.merge(df2,df_user, left_on='userid', right_on='userid', how='inner')
df3
```

```python
df3.sort_values(by=['shopid','event_time']).reset_index(drop=True)
```

```python
shopcount1 = df3.groupby(['shopid'],as_index=False)['shopid'].agg({'cnt':'count'}) #as_index:讓shopid不要變成index
shopcount1
```

```python
df_shop = shopcount1[shopcount1['cnt'] > 2].reset_index(drop=True)
del df_shop['cnt']
df_shop
```

```python
df4 = pd.merge(df3,df_shop, left_on='shopid', right_on='shopid', how='inner') 
df4
```

```python
df5 = df4.groupby(['shopid','userid'],as_index=False)['userid'].agg({'cnt':'count'})
df5
```

```python
df6 = df5[df5['cnt'] > 2].reset_index(drop=True)
df6 = df6.sort_values(by=['shopid','cnt'],ascending = False).reset_index(drop=True)
df6
```

```python
n = len(df6)
i_list = []
x = 0
temp = df6.iloc[0,0]
temp_count = df6.iloc[0,2]
for i in range(n):
    if df6.iloc[i,0] == temp and df6.iloc[i,2] < temp_count:
        i_list.append(i)
    elif df6.iloc[i,0] != temp:
        temp = df6.iloc[i,0]
        temp_count = df6.iloc[i,2]
for i in range(len(i_list)):
    df6 = df6.drop(df6.index[[i_list[i]-x]])
    x = x+1
```

```python
id_list = []
for i in range(len(df6)):
    if df6.iloc[i, 0] not in id_list:
        id_list.append(df6.iloc[i, 0])
        
del_list = []
for j in range(len(df)):
    if df.iloc[j, 1] not in id_list:
        del_list.append(j)
        
df7 = df
y = 0
for k in range(len(del_list)):
    df7 = df7.drop(df7.index[[del_list[k]-y]])
    y = y + 1
```

```python
df8 = df7.sort_values(by=['shopid','event_time']).reset_index(drop=True)
df8
```

```python
from datetime import datetime, timedelta
dateFormatter = "%Y-%m-%d %H:%M:%S"
temp_list = []
final = []

for i in range(0,len(df8)):
    temp_shopid = df8.iloc [i, 1]
    temp_event_time = df8.iloc [i, 3]
    count = 0
    temp_list = []
    for j in range(i,len(df8)):
        if df8.iloc [j, 1] == temp_shopid and (datetime.strptime(df8.iloc [j, 3], dateFormatter) - datetime.strptime(temp_event_time, dateFormatter)) <= timedelta(hours=1):
            if df8.iloc [j, 2] not in temp_list:
                count += 1
            temp_list.append(df8.iloc [j, 2])
        

        elif df8.iloc [j, 1] != temp_shopid or (datetime.strptime(df8.iloc [j, 3], dateFormatter) - datetime.strptime(temp_event_time, dateFormatter)) > timedelta(hours=1):
            if len(temp_list)/count >= 3:
                set_list = list(set(temp_list))
                for k in range(len(set_list)):
                    if temp_list.count(set_list[k]) >= 3 and [temp_shopid, set_list[k], temp_list.count(set_list[k])] not in final:
                        final.append([temp_shopid, set_list[k] ,temp_list.count(set_list[k])])
 
            break
```

```python
data1 = data.sort_values(by=[0,2],ascending = False).reset_index(drop=True)
data1
```

```python
data2 = data1.rename(columns = {0:'shopid', 1:'userid', 2:'cnt'})
data2
```

```python
temp_shopid = data2.iloc[0, 0]
temp_count = data2.iloc[0, 2]
final_del_list = []

for i in range(len(data2)):
    if data2.iloc[i, 0] == temp_shopid and data2.iloc[i, 2] < temp_count:
        final_del_list.append(i)
    elif data2.iloc[i, 0] != temp_shopid:
        temp_shopid = data2.iloc[i, 0]
        temp_count = data2.iloc[i, 2]

z =  0
for j in range(len(final_del_list)):
    data2 = data2.drop(data2.index[[final_del_list[j]-z]])
    z = z + 1

final = data2.reset_index(drop=True)
final
```

```python
same_list = []

for i in range(len(final)-1):
    if final.iloc[i, 0] == final.iloc[i+1, 0]:
        same_list.append(i+1)
        final.iloc[i,1] = str(final.iloc[i,1]) + "&" + str(final.iloc[i+1,1])


w = 0 
for j in range(len(same_list)):
    final= final.drop(final.index[[same_list[j]-w]])
    w = w + 1

final = final.reset_index(drop=True)
final
```

```python
delete_list = []
for i in range(len(final)):
    if final.iloc[i, 0] not in delete_list:
        delete_list.append(final.iloc[i, 0])
        
del_list = []
for j in range(len(df)):
    if df.iloc[j, 1] in delete_list:
        del_list.append(j)

df_final = df
y = 0
for k in range(len(del_list)):
    df_final = df_final.drop(df_final.index[[del_list[k]-y]])
    y = y + 1

df_final['userid_final']=0
```

```python
del df_final['orderid']
del df_final['userid']
del df_final['event_time']
del df_final['year']
del df_final['month']
del df_final['day']
del df_final['hour']
del df_final['mins']
```

```python
df_final = df_final.groupby('shopid',as_index =False).sum()
df_final = df_final.rename(columns = {'userid_final':'userid'})
df_final
```

```python
submission = pd.merge(final,df_final, left_on=['shopid','userid'], right_on=['shopid','userid'], how='outer')
submission
submission.to_csv('competition1_submission.csv', index = False)
```
